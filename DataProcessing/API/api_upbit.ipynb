```python
mycoin = upbit.get_weeks_candles('KRW-EOS', count=365)
mycoin[-1]
```

```python
high_price = 0
for coin in mycoin:
    if high_price < coin['high_price']:
        date = coin['candle_date_time_kst']
        high_price = coin['high_price']

print(date, high_price)
```
